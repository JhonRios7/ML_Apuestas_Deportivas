Limpieza de los datos

```python
matches.dtypes #observar datos categoricos y convertirlos en númericos para el algoritmo de ML
```

```python
del matches["Comp"]
```

```python
del matches["Notes"]
```

```python
#Convertir 'Date' en tipo de datos fecha y hora .. no quedaria como objeto
matches["Date"] = pd.to_datetime(matches["Date"])
```

```python
#convertir los registros de la columna 'Venue' (local:1 o visitante:0) en datos númericos en una nueva columna
matches["Venue_code"] = matches["Venue"].astype("category").cat.codes
```

```python
#Codigo para cada equipo oponente
matches["Opp_code"] = matches["Opponent"].astype("category").cat.codes
```

```python
#Codificar los dias de la semana
matches["Day_code"] = matches["Date"].dt.dayofweek
```

```python
matches["Hour"] = matches["Time"].str.replace(":.+", "", regex=True).astype("int") #solo dejar la hora (puede ser que algunos equipos jueguen mejor a cierta hora) NÚMERO ENTERO
```

```python
#Devuelve valor booleano, si el resultado es W: VERDADERO(1) ... Y SI NO ES W:FALSO(0)
matches["Target"] = (matches["Result"] == "W").astype("int")
```

Creación modelo de ML

```python
from sklearn.ensemble import RandomForestClassifier
```

Inicializaremos el modelo con algunos parametros:
- n_estimators : número de arboles de decisión individuales que queremos que entrene /entre mas alto este valor mas tardara el algoritmo para ejecutar pero mejor sera mas preciso.
- min_simples_split: división minima de muestras que queremos tener en una hoja del arbold e decisión anets de dividir el nodo, cuanto mayor sea es menos probable que haya sobreajuste, pero menor sera su precision en los datos de train.
- random_state = parametros aleatorios del bosque

```python
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

División Datos de Train & Test

```python
train = matches[matches["Date"] < '2024-01-01']
```

```python
test = matches[matches["Date"] > '2024-01-01']
```

```python
predictors = ["Venue_code", "Opp_code", "Hour", "Day_code"] #Variables predictoras
```

```python
rf.fit(train[predictors], train["Target"]) #Ajustamos modelo usando predictores de train para predecir objetivo
```

```python
preds = rf.predict(test[predictors]) #Predicción
```

Mejorar la precisión con medias móviles 
- Crear mas predictores 
Dividiremos el cuadro de datos de nuestros partidos por equipo ya que queremos calcular promedios moviles para el rendimiento del equipo (ultimos juegos, tiros al arco, goles en contra, etc)

```python
#creamos la variable de partidos agrupados para ver los ultimos partidos
grouped_matches = matches.groupby("Team")
```

```python
group = grouped_matches.get_group("Manchester City").sort_values("Date") 
```

```python
#creamos una función 'promedios moviles' : toma conjunto de columnas para las que se quiere calcular los
#promedios moviles de las ultimas 3 semanas(juegos) y luego incluira un conjunto de columnas nuevas a las que queremos asignar los promedios
#moviles .
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("Date") #ordenar grupo por fecha ASCENDENTE
    rolling_stats = group[cols].rolling(3, closed='left').mean() #calculo prom.movil de esas columnas 
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)#eliminar valores faltantes
    return group
```

```python
cols = ["GF", "GA", "Sh", "SoT", "Dist", "FK", "PK", "PKatt"]
new_cols = [f"{c}_rolling" for c in cols]

rolling_averages(group, cols, new_cols)
```

```python
#Aplicar esto a todos los matches  (el anterior era solo para el Man City) .. aplica la función a cada equipo
matches_rolling = matches.groupby("Team").apply(lambda x: rolling_averages(x, cols, new_cols))
```

Reentrenamiento del modelo de aprendizaje automático 

```python
#crearemos una función que nos facilite seguir iterando nuestro algoritmo porque no tenemos que seguir
#repitiendo el codigo de prueba de train
def make_predictions(data, predictors):
    train = data[data["Date"] < '2024-01-01']
    test = data[data["Date"] > '2024-01-01']
    rf.fit(train[predictors], train["Target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["Target"], predicted=preds), index=test.index)
    error = precision_score(test["Target"], preds)
    return combined, error
```

```python
combined, error = make_predictions(matches_rolling, predictors + new_cols)
```

Combinación de pronostico como local y visitante

```python
#normalizar los nombres para que sean iguales en Team y en Opponent para algunos equipos /creamos diccionario
class MissingDict(dict):
    __missing__ = lambda self, key: key

map_values = {"Brighton and Hove Albion": "Brighton", "Manchester United": "Manchester Utd", "Newcastle United": "Newcastle Utd", "Tottenham Hotspur": "Tottenham", "West Ham United": "West Ham", "Wolverhampton Wanderers": "Wolves"} 
mapping = MissingDict(**map_values)
```

```python
combined["new_team"] = combined["Team"].map(mapping) 
```

```python
#FUSIONAR el marco de datos consigo mismo
merged = combined.merge(combined, left_on=["Date", "new_team"], right_on=["Date", "Opponent"])
```
