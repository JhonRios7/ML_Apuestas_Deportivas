```python
import requests #solicitud para el website
```

```python
standings_url =  "https://fbref.com/en/comps/9/Premier-League-Stats" #Definimos la URL con la que vamos a comenzar
```

```python
data = requests.get(standings_url) #Se descarga la pág con el método .get en la libreria requests // hace la solicitud al servidor y descarga el html
```

```python
data.text
```

Convertimos esta cadena html en elementos graficos que realmente se puedan ver y entender.

```python
from bs4 import BeautifulSoup #para extraer información de una página web y convertirla en un formato estructurado como un DataFrame
```

```python
soup = BeautifulSoup(data.text) #Inicializamos el objeto soup
standings_table = soup.select('table.stats_table')[0] #seleccionamos la etiqueta de la tabla 
links = standings_table.find_all('a') #encontrar todas las etiquetas en la tabla
links = [l.get("href") for l in links] #obtener la propiedad 'href' de cada enlace  .... l=lista de comprensión ... recorre todos los elementos y encuentra el valor de la propiedad
links = [l for l in links if '/squads/' in l] #filtrat los enlaces para tener los enlaces solo del escuadron .. con otra lista de comprensión
```

```python
links
```

Se puede visualizar que muestra la ultima parte del enlace y no tienen el comienzo con el dominio, asi que se convierten los enlaces a URL Completas. usando CADENA DE FORMATO

```python
team_urls = [f"https://fbref.com{l}" for l in links] #Toma cada uno de nuestros enlaces y agrega esta cadena al principio
```

```python
team_urls
```

```python
data = requests.get(team_urls[0])
```

Extracción de estadísticas de partidos utilizando pandas y requests

```python
import pandas as pd
matches = pd.read_html(data.text, match="Scores & Fixtures")[0]
```

```python
matches
```

```python
# Convertir la columna 'Date' a tipo datetime
matches['Date'] = pd.to_datetime(matches['Date'])

# Encontrar la fecha más antigua y la fecha más reciente
fecha_mas_antigua = matches['Date'].min().strftime('%Y-%m-%d')
fecha_mas_reciente = matches['Date'].max().strftime('%Y-%m-%d')

print("La fecha más antigua es:", fecha_mas_antigua)
print("La fecha más reciente es:", fecha_mas_reciente)
```

obtener estadísticas de tiro en partido, con las caracteristicas de las temporadas pasadas tambien atravez de un ciclo

```python
years = list(range(2024,2022, -1))
```

```python
all_matches = []
```

```python
import time
for year in years:
    data = requests.get(standings_url) 
    soup = BeautifulSoup(data.text) #analizar el html
    standings_table = soup.select('table.stats_table')[0] #tabal estadisticas de puntos (equipo individual)

    links = [l.get("href") for l in standings_table.find_all('a')]#para encontarr todos los enlaces de los equipos
    links = [l for l in links if '/squads/' in l]
    team_urls = [f"https://fbref.com{l}" for l in links]
    
    previous_season = soup.select("a.prev")[0].get("href") #obtener la season anteriro
    standings_url = f"https://fbref.com{previous_season}" #y luego se convertida en una URL absoluta
    
    for team_url in team_urls: #recorrer las URL del equipo
        team_name = team_url.split("/")[-1].replace("-Stats", "").replace("-", " ") #eliminar individualmente los registros de partidos de cada equipo
        data = requests.get(team_url) #URL del equipo que nos permiyte obtener tabla de resultados y partidos
        matches = pd.read_html(data.text, match="Scores & Fixtures")[0] #Primero se analiza tabla de puntajes y partidos
        soup = BeautifulSoup(data.text) #sacamos todos los enlaces..
        links = [l.get("href") for l in soup.find_all('a')] #.. y los utilizamos para obtener el enlace a la pág de tiro
        links = [l for l in links if l and 'all_comps/shooting/' in l]
        data = requests.get(f"https://fbref.com{links[0]}") # se convierten a URL absolutas
        shooting = pd.read_html(data.text, match="Shooting")[0] #ller estadisticas de disparo
        shooting.columns = shooting.columns.droplevel() #eliminar indice superior en tabla
        try: #fusionar estadisticas de tiro con las estadisticas de partido
            team_data = matches.merge(shooting[["Date", "Sh", "SoT", "Dist", "FK", "PK", "PKatt"]], on="Date")
        except ValueError: #cuando las estadisticas de tiro de un equipo este vacio, y de error simplemente continue con el bucle (nos saltamos los equipos dodne no esten disponibles estas estadisticas)
            continue
        team_data = team_data[team_data["Comp"] == "Premier League"] #competición premier league
        
        team_data["Season"] = year
        team_data["Team"] = team_name
        all_matches.append(team_data)
        time.sleep(1)
```

time.sleep(1) --> no se hace demasiado rapido el scraping porque puede ralentizar el sitio web y dificulta el funcionamiento eficaz de este, por lo que al reducir la velocidad con la que estamos scrapiando nos aseguramos de no hacerlo.

```python
match_df = pd.concat(all_matches) #concatenamos todos nuestros macros de datos individuales en un solo macro
```

Sh: tiros, SoT: Tiros a puerta

```python
match_df
```

```python
# Convertir la columna 'Date' a tipo datetime
match_df['Date'] = pd.to_datetime(match_df['Date'])

# Encontrar la fecha más antigua y la fecha más reciente
fecha_mas_antigua = match_df['Date'].min().strftime('%Y-%m-%d')
fecha_mas_reciente = match_df['Date'].max().strftime('%Y-%m-%d')

print("La fecha más antigua es:", fecha_mas_antigua)
print("La fecha más reciente es:", fecha_mas_reciente)
```
